```python
from PIL import Image
import numpy as np
from scipy.ndimage import rotate
import matplotlib.pyplot as plt

# Load an image using PIL
image_path = '/content/Flower Color Images Dataset/flowers/flowers/19_010.png'
original_image = Image.open(image_path)

def rotate_image_90_degrees(image):
    return rotate(image, 90, reshape=False, mode='nearest')

image_array = np.array(original_image)

# Поверни изображение
rotated_image = rotate_image_90_degrees(image_array)

# Отобрази результат
plt.imshow(rotated_image)
plt.axis('off')
plt.show()
```
