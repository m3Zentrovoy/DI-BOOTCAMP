```python

import pandas as pd
df = pd.read_csv('/Users/zentrovoy/Documents/BOOTCAMP/W3/D3/Exercise/datascience_salaries.csv')
df
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['salary_normalized'] = scaler.fit_transform(df[['salary']])
df
```

```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np

numerical_cols = df.select_dtypes(include=[np.number]).columns
df_numerical = df[numerical_cols]
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(df_numerical)
df_reduced = pd.DataFrame(reduced_data, columns=['PC1', 'PC2'])

print(display((df_reduced.head(10))))
```

```python
import pandas as pd

grouped = df.groupby('experience_level')
mean_df = grouped[['salary']].mean().round(2)
median_df = grouped[['salary']].median().round(2)

mean_df.columns = ['Mean Salary']
median_df.columns = ['Median Salary']

print(mean_df.sort_values(by='Mean Salary'))
print(median_df.sort_values(by='Median Salary'))
```
