```python
#EX4
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a synthetic dataset for students' CGPA (since we don't have the actual data)
np.random.seed(42)  # For reproducibility
cgpa_data = np.random.normal(loc=7.5, scale=1.0, size=1000)  # CGPA: mean=7.5, std=1.0
cgpa_data = np.clip(cgpa_data, 0, 10)  # Ensure CGPA is between 0 and 10
df = pd.DataFrame({'CGPA': cgpa_data})

# Create a histogram of CGPA using Seaborn's histplot
sns.histplot(df['CGPA'], bins=20, color='skyblue')

# Customize the histogram
plt.title('Distribution of Students CGPA')
plt.xlabel('CGPA')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
#EX 5

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create synthetic dataset based on research findings
np.random.seed(42)  # For reproducibility
# Men: Lower anxiety levels (mean=3, std=1.5)
men_anxiety = np.random.normal(loc=3, scale=1.5, size=500)
# Women: Higher anxiety levels (mean=5, std=1.8)
women_anxiety = np.random.normal(loc=5, scale=1.8, size=500)

# Create DataFrame
df = pd.DataFrame({
    'Anxiety_Level': np.concatenate([men_anxiety, women_anxiety]),
    'Gender': ['Men'] * 500 + ['Women'] * 500
})

# Create histogram using Seaborn's histplot
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Anxiety_Level', hue='Gender', bins=30, palette={'Men': 'blue', 'Women': 'pink'}, alpha=0.6)

# Customize the plot
plt.title('Distribution of Anxiety Levels Across Genders')
plt.xlabel('Anxiety Level (Scale 0-10)')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming df is already created from previous steps
np.random.seed(42)  # For reproducibility
ages = np.random.normal(loc=40, scale=15, size=1000)
ages = np.clip(ages, 18, 80)
ages = np.round(ages).astype(int)
panic_attacks = []
for age in ages:
    if 30 <= age <= 50:
        has_panic = np.random.choice(['Yes', 'No'], p=[0.3, 0.7])
    else:
        has_panic = np.random.choice(['Yes', 'No'], p=[0.15, 0.85])
    panic_attacks.append(has_panic)
df = pd.DataFrame({'Age': ages, 'Do you have Panic Attacks?': panic_attacks})

# Convert panic attack responses to numeric values (Yes=1, No=0)
df['Panic Attacks Numeric'] = df['Do you have Panic Attacks?'].map({'Yes': 1, 'No': 0})

# Set plot aesthetics using Seaborn's customization
sns.set_context('talk')  # Adjust context for better readability
sns.set_style('whitegrid')  # Set background style to whitegrid

# Create a custom color palette from light pink to dark purple
palette = sns.color_palette("blend:#FF9999,#4B0082", as_cmap=True)

# Create a scatter plot using sns.scatterplot with gradient color based on Age
plt.figure(figsize=(10, 5))  # Set figure size
sns.scatterplot(x='Age', y='Panic Attacks Numeric', hue='Age', data=df, palette=palette, s=100, alpha=0.6)

# Customize the plot
plt.title('Relationship Between Age and Panic Attacks')
plt.xlabel('Age')
plt.ylabel('Panic Attacks (1 = Yes, 0 = No)')

# Display the plot
plt.show()
```
