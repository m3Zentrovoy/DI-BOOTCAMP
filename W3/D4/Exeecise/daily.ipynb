```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns



df = pd.read_csv('/Users/zentrovoy/Documents/BOOTCAMP/W3/D4/Exeecise/2019.csv')
df = df.set_index('Overall rank')
# df = df.sort_values(by = 'Country or region', ascending=True)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.jointplot(x='Score', y='Social support', data=df, kind='reg', color='orange')
plt.show()
```

```python
top_countries = df.sort_values(by='Score', ascending=False).head(10)

# Create subplots: two vertically stacked plots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Plot 1: GDP per Capita by Country
ax1.bar(top_countries['Country or region'], top_countries['GDP per capita'], color='skyblue')
ax1.set_title('GDP per Capita for Top 10 Happiest Countries')
ax1.set_xlabel('Country')
ax1.set_ylabel('GDP per Capita')
ax1.tick_params(axis='x', rotation=45)

# Plot 2: Healthy Life Expectancy by Country
ax2.bar(top_countries['Country or region'], top_countries['Healthy life expectancy'], color='lightgreen')
ax2.set_title('Healthy Life Expectancy for Top 10 Happiest Countries')
ax2.set_xlabel('Country')
ax2.set_ylabel('Healthy Life Expectancy')
ax2.tick_params(axis='x', rotation=45)

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Select top 10 countries based on Score for better readability
top_countries = df.sort_values(by='Score', ascending=False).head(10)

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot 1: Bar plot for GDP per Capita
ax1.bar(top_countries['Country or region'], top_countries['GDP per capita'], color='skyblue', label='GDP per Capita')
ax1.set_xlabel('Country')
ax1.set_ylabel('GDP per Capita', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.tick_params(axis='x', rotation=45)

# Create a second y-axis for the line plot
ax2 = ax1.twinx()
ax2.plot(top_countries['Country or region'], top_countries['Healthy life expectancy'], color='green', marker='o', label='Healthy Life Expectancy')
ax2.set_ylabel('Healthy Life Expectancy', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Add a title
plt.title('Comparison of GDP per Capita and Healthy Life Expectancy (Top 10 Happiest Countries)')

# Add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```
