```python
import matplotlib.pyplot as plt
import numpy as np


# Sample Data
x = np.linspace(0, 100, 50)
y = 2*x+ (100*np.random.rand(50))

# Create Scatter Plot
plt.scatter(x, y)

# Calculate Trend Line (Linear)
coefficients = np.polyfit(x, y, 1)
trend_line = np.poly1d(coefficients)

# Generate Trend Line Points
trend_line_y = trend_line(x)

# Plot Trend Line
plt.plot(x, trend_line_y, 'r--')

# Show Plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
data = {
    'Apple': 50,
    'Banana': 25,
    'Orange': 15,
    'Grapes': 30,
    'Peach': 10,
    'Cherry': 20,
    'Mango': 40
}

# Sort and get top 5 values
top_items = sorted(data.items(), key=lambda x: x[1], reverse=True)[:5]
labels, values = zip(*top_items)

# Create pie chart
plt.figure(figsize=(6,6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Fruit Quantities')
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle.
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample data
data = sns.load_dataset('tips')
sns.histplot(data = data, x='total_bill', bins=20)
plt.title('Histogram of Total Bill')
plt.show()
```

```python
sns.kdeplot(data['total_bill'], shade=True, color='purple')
plt.title('KDE of Total Bill')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=data, kind='scatter', color='green', hue='smoker')
plt.show()
```

```python
iris = sns.load_dataset('iris')
g = sns.PairGrid(iris, hue='species')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, cmap='Blues')
g.map_diag(sns.histplot, color='teal')
g.add_legend()
plt.show()
```

```python
correlation_matrix = iris.drop(['species'], axis=1).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
