```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cities = [
    'New York', 'Los Angeles', 'Tokyo', 'London', 'Sydney',
    'Cape Town', 'Moscow', 'Dubai', 'Toronto', 'Sao Paulo'
]
months = [
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
]

temperatures = np.random.uniform(low=-5, high=35, size=(10, 12))
temperatures = np.round(temperatures, 1)
df = pd.DataFrame(temperatures, index=cities, columns=months)
df

```

```python
# 2. Print annual average temperature for each city
print(f'Annual average \n{df.mean(axis=1)}')
```

```python
# 3. Find the city with the highest average temperature per month
print(f'City with the highest average \n{df.idxmax()}')
```

```python

# 4. Find the city with the lowest average temperature per month
print(f'City with the lowest average \n{df.idxmin()}')
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
for city in cities:
    plt.plot(months, df.loc[city], marker='o', label=city)
plt.title('Monthly Temperature Trends for 10 Cities')
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.grid(True)
plt.tight_layout()
```

```python
plt.subplot(2, 1, 2)
annual_avg = df.mean(axis=1)
bars = plt.bar(cities, annual_avg, color='skyblue')

bars[cities.index(df.mean(axis=1).idxmax())].set_color('red')  
bars[cities.index(df.mean(axis=1).idxmin())].set_color('blue') 
plt.title('Annual Average Temperature by City')
plt.xlabel('City')
plt.ylabel('Average Temperature (C)')
plt.xticks(rotation=45)
plt.grid(True, axis='y')

plt.tight_layout()
plt.show()
```

**Data Overview**: Generated synthetic monthly temperatures for 10 cities over 12 months (range: -5C to 35C).
- **Annual Averages**: Calculated average yearly temperatures for each city.
- **Highest Average**: Identified Tokyo with the highest average temperature per month.
- **Lowest Average**: Identified Los Angeles with the lowest average temperature per month.

## Summary of Temperature Analysis

This analysis examines synthetic monthly temperatures for 10 cities over 12 months in 2025, revealing notable trends:

- **Highest Average Temperature**: Tokyo, with an annual average of 22.1C, shows extreme monthly swings, e.g., from 3.2C in Feb to 31.6C in Nov.
- **Lowest Average Temperature**: Los Angeles, with an annual average of 11.8C, also displays drastic fluctuations, e.g., from 0.8C in Apr to 33.1C in May.
- **Unrealistic Monthly Swings**: Most cities exhibit improbable temperature jumps each month. For instance, Moscow drops to -2.4C in Jun but rises to 31.0C in Dec, and Dubai fluctuates from 4.5C in May to 34.5C in Jun, highlighting the synthetic nature of the data.
